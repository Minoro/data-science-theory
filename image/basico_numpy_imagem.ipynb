# Imagens com Numpy

Exemplos de código para manipulação básica de imagens com numpy.

Normalmente em produção utiliza-se bibliotécas de mais alto nível, como opencv ou simplecv, mas esses exemplos podem ser utilizados como base para manipulação de matrizes com outros tipos de dados.

```python
import numpy as np
from matplotlib import pyplot as plt
import scipy

face = scipy.misc.face() 

#valores mínimo, máximo e média
print('Valo mínimo: ', face.min())
print('Valor máximo: ', face.max())
print('Valor médio: ', face.mean())

#exibindo imagem base de exemplo
plt.imshow(face)
```

# Manipulando indices da imagem

Uma imagem pode ser manipulada pelas operações sobre array do numpy.

Uma imagem RGB é representada por uma matriz tri-dimensional, no formato: $image[x][y][rgb\_value]$.
Assim ao acessar $image[0][0]$ recupera-se a tupla $rgb\_value$ com a cor o primeiro pixel da imagem.

Também é possível recortar uma imagem por meio de `indice slicing` sobre o array original.


```python
print("Valor do primeiro pixel")
print(face[0][0])


sub_image = face[200:500, 400:900]

plt.subplot(1,2,1)
plt.title("Sub-imagem")
plt.imshow(sub_image)



r_chanel = sub_image[:,:, 0]
print("Tipo de dado do canel vermelho: ", r_chanel.dtype)
plt.subplot(1,2,2)
plt.title("Canal vermelho")
plt.imshow(r_chanel, cmap='gray')

```

```python
gray_image = face.dot([0.07, 0.72, 0.21])

gray_image = gray_image.astype(np.uint8)

print(gray_image.max())
print(gray_image.shape)

plt.title("Imagem em tons de cinza")
plt.imshow(gray_image, cmap='gray')
```
